```python
# Your code here

# Assuming your DataFrame is named 'df'
wholesale['Channel'] = wholesale['Channel'].astype('category')
wholesale['Region'] = wholesale['Region'].astype('category')
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, RobustScaler


# Select highly correlated features
high_corr_features = wholesale[['Milk', 'Grocery', 'Detergents_Paper']]

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(high_corr_features)

# Apply PCA
pca = PCA(n_components=1)
pca_result = pca.fit_transform(scaled_data)

# Replace the original features with the PCA result
wholesale['Milk_Grocery_Detergents_PCA'] = pca_result
wholesale.drop(['Milk', 'Grocery', 'Detergents_Paper'], axis=1, inplace=True)
```

```python
# Select features with outliers
features_with_outliers = ['Fresh', 'Milk_Grocery_Detergents_PCA', 'Frozen', 'Delicassen']

# Apply Robust Scaling
scaler = RobustScaler()
wholesale[features_with_outliers] = scaler.fit_transform(wholesale[features_with_outliers])
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Range of cluster numbers to try
cluster_range = range(1, 11)  # Try 1 to 10 clusters

# List to store the WCSS for each number of clusters
wcss = []

for k in cluster_range:
    # Initialize and fit KMeans
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customers_scale)
    # Append WCSS to the list
    wcss.append(kmeans.inertia_)

# Plotting the Elbow Curve
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, wcss, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```

### Looking to the elbow we can choose 4 like the correct number of clusters

```python
from sklearn.cluster import KMeans

# Your code here:

# Assuming 'customers_scale' is your scaled DataFrame
# Decide on the number of clusters, 4 is choosen
n_clusters = 4

# Initialize the KMeans model
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Fit the model to the scaled data
kmeans.fit(customers_scale)

# Extract labels and add them to the original DataFrame
wholesale['labels'] = kmeans.labels_

# Display the first few rows of the DataFrame with cluster labels
print(wholesale.head())
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN

# Your code here


# Initialize DBSCAN model
# `eps` is the maximum distance between two samples for them to be considered as in the same neighborhood.
dbscan = DBSCAN(eps=0.5)

# Fit the model to your scaled dataset
# This finds clusters within your data based on density
dbscan.fit(customers_scale)

# Extract the cluster labels which are automatically generated once the model is fitted
# `labels_` is an attribute containing the predicted cluster labels for each sample
labels_dbscan = dbscan.labels_

# Add these labels to the original wholesale DataFrame
# Create a new column in your DataFrame to store these labels
wholesale['labels_DBSCAN'] = labels_dbscan

# Display the first few rows of the DataFrame to ensure that the new column has been added correctly
print(wholesale.head())

# For further analysis, count how many samples are in each DBSCAN cluster, including noise if any (-1 label)
dbscan_label_counts = wholesale['labels_DBSCAN'].value_counts()
print("DBSCAN Cluster Counts:")
print(dbscan_label_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Confirm column names first
print("Available Columns: ", wholesale.columns)

# Verify and confirm that the columns exist in your DataFrame
# Update these names according to your DataFrame if needed
wholesale.columns = wholesale.columns.str.strip()  # Strip any extra whitespace

def plot_comparison(x, y, title):
    plt.figure(figsize=(14, 6))

    # K-Means plot: Check the column names and ensure they exist in the data
    plt.subplot(1, 2, 1)
    sns.scatterplot(x=x, y=y, hue='labels', data=wholesale, palette='viridis', legend='full')
    plt.title(f'K-Means Clustering: {title}')
    plt.xlabel(x)
    plt.ylabel(y)

    # DBSCAN plot: Similarly, confirm column names here too
    plt.subplot(1, 2, 2)
    sns.scatterplot(x=x, y=y, hue='labels_DBSCAN', data=wholesale, palette='viridis', legend='full')
    plt.title(f'DBSCAN Clustering: {title}')
    plt.xlabel(x)
    plt.ylabel(y)

    plt.tight_layout()
    plt.show()

# Ensure column names match the DataFrame
if 'Detergents_Paper' in wholesale.columns and 'Milk' in wholesale.columns:
    plot_comparison('Detergents_Paper', 'Milk', 'Detergents_Paper vs Milk')

if 'Grocery' in wholesale.columns and 'Fresh' in wholesale.columns:
    plot_comparison('Grocery', 'Fresh', 'Grocery vs Fresh')

if 'Frozen' in wholesale.columns and 'Delicassen' in wholesale.columns:
    plot_comparison('Frozen', 'Delicassen', 'Frozen vs Delicassen')
```
